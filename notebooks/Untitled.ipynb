````python
"""
╔══════════════════════════════════════════════════════════════════╗
║           WHILE-SLEEPING — TECH STACK SETUP SCRIPT              ║
║           Barclays Self-Healing Journey System                   ║
║           Decision Intelligence — Real-Time Pain Detection       ║
╚══════════════════════════════════════════════════════════════════╝

Run this in JupyterLab to install and verify your full stack.
"""

import subprocess
import sys
import os

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 1: CORE DATA STACK
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
core_packages = [
    "pyspark",           # Local Databricks engine
    "delta-spark",       # Delta Lake tables
    "pandas",            # Data manipulation
    "numpy",             # Numerical computing
    "pyarrow",           # Parquet file format
    "fastparquet",       # Parquet support
    "boto3",             # S3 / MinIO connection
    "python-dotenv",     # Manage credentials safely
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 2: STREAMING & EVENTS
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
streaming_packages = [
    "kafka-python",      # Kafka event streaming
    "confluent-kafka",   # Confluent Kafka client
    "faust-streaming",   # Stream processing in Python
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 3: AGENT FRAMEWORK
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
agent_packages = [
    "langchain",                  # Agent tools framework
    "langgraph",                  # Agent workflow orchestration
    "langchain-community",        # Community integrations
    "langchain-core",             # Core abstractions
    "langchain-ollama",           # Qwen via Ollama
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 4: MCP (NERVOUS SYSTEM)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
mcp_packages = [
    "mcp",               # Model Context Protocol SDK
    "httpx",             # Async HTTP for MCP servers
    "fastapi",           # MCP server framework
    "uvicorn",           # ASGI server for FastAPI
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 5: AI & LLM
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
ai_packages = [
    "anthropic",         # Claude API (production path)
    "ollama",            # Qwen local LLM
    "tiktoken",          # Token counting
    "sentence-transformers",  # Embeddings for memory
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 6: ANALYTICS & DETECTION
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
analytics_packages = [
    "scikit-learn",      # ML models & anomaly detection
    "scipy",             # Statistical analysis
    "statsmodels",       # Time series & baselines
    "prophet",           # Forecasting & trend detection
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 7: DASHBOARD
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
dashboard_packages = [
    "streamlit",         # Main dashboard framework
    "plotly",            # Interactive charts
    "altair",            # Declarative visualisation
    "matplotlib",        # Static charts
    "seaborn",           # Statistical visualisation
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 8: DATA QUALITY & LINEAGE
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
quality_packages = [
    "great-expectations", # Data quality checks
    "dbt-core",           # Semantic layer & transforms
    "sqlalchemy",         # Database connections
    "alembic",            # Schema migrations
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 9: NOTIFICATIONS (Human Guardrail Alerts)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
notification_packages = [
    "twilio",            # WhatsApp & SMS alerts
    "slack-sdk",         # Slack notifications
    "requests",          # HTTP requests
    "aiohttp",           # Async HTTP
]

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# LAYER 10: UTILITIES
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
utility_packages = [
    "pydantic",          # Data validation
    "loguru",            # Advanced logging
    "rich",              # Beautiful terminal output
    "typer",             # CLI tools
    "schedule",          # Task scheduling
    "uuid",              # Unique IDs
    "arrow",             # Date/time handling
]


def install_package(package):
    """Install a single package and return success status."""
    try:
        subprocess.check_call(
            [sys.executable, "-m", "pip", "install", package, "--quiet"],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL
        )
        return True
    except subprocess.CalledProcessError:
        return False


def install_layer(layer_name, packages, emoji):
    """Install all packages in a layer."""
    print(f"\n{emoji} Installing {layer_name}...")
    print("─" * 50)
    
    success = []
    failed = []
    
    for package in packages:
        result = install_package(package)
        if result:
            print(f"  ✅ {package}")
            success.append(package)
        else:
            print(f"  ❌ {package} — failed")
            failed.append(package)
    
    return success, failed


def verify_stack():
    """Verify all critical packages are importable."""
    print("\n\n🔍 VERIFYING STACK...")
    print("═" * 50)
    
    critical_checks = {
        "pyspark":        "PySpark (Databricks engine)",
        "delta":          "Delta Lake (storage)",
        "pandas":         "Pandas (data manipulation)",
        "numpy":          "NumPy (numerical computing)",
        "pyarrow":        "PyArrow (Parquet format)",
        "langchain":      "LangChain (agent tools)",
        "langgraph":      "LangGraph (agent orchestration)",
        "mcp":            "MCP (nervous system)",
        "streamlit":      "Streamlit (dashboard)",
        "plotly":         "Plotly (charts)",
        "sklearn":        "Scikit-learn (ML/anomaly detection)",
        "anthropic":      "Anthropic (Claude API)",
        "sqlalchemy":     "SQLAlchemy (database connections)",
        "fastapi":        "FastAPI (MCP servers)",
        "kafka":          "Kafka (event streaming)",
    }
    
    all_good = True
    for module, description in critical_checks.items():
        try:
            __import__(module)
            print(f"  ✅ {description}")
        except ImportError:
            print(f"  ⚠️  {description} — not available")
            all_good = False
    
    return all_good


def create_project_structure():
    """Create the while-sleeping project folder structure."""
    print("\n\n📁 CREATING PROJECT STRUCTURE...")
    print("═" * 50)
    
    folders = [
        "data/raw",
        "data/delta/raw_events",
        "data/delta/cleaned_events",
        "data/delta/pain_scores",
        "data/delta/aggregations",
        "data/delta/hypothesis_library",
        "data/delta/agent_memory",
        "data/delta/audit_log",
        "data/delta/outcomes",
        "src/simulate",
        "src/pipeline",
        "src/detection",
        "src/agents",
        "src/mcp_servers",
        "src/hypothesis",
        "src/remediation",
        "src/dashboard",
        "src/notifications",
        "notebooks",
        "tests",
        "config",
        "docs",
    ]
    
    for folder in folders:
        os.makedirs(folder, exist_ok=True)
        print(f"  📂 {folder}")
    
    # Create placeholder files
    placeholders = [
        "src/__init__.py",
        "src/simulate/__init__.py",
        "src/pipeline/__init__.py",
        "src/detection/__init__.py",
        "src/agents/__init__.py",
        "src/mcp_servers/__init__.py",
        "src/hypothesis/__init__.py",
        "src/remediation/__init__.py",
        "src/dashboard/__init__.py",
        "src/notifications/__init__.py",
        "config/settings.py",
    ]
    
    for filepath in placeholders:
        if not os.path.exists(filepath):
            with open(filepath, 'w') as f:
                f.write(f"# while-sleeping — {filepath}\n")
    
    print("\n  ✅ Project structure created!")


def print_summary(all_results):
    """Print final installation summary."""
    print("\n\n" + "═" * 50)
    print("📊 INSTALLATION SUMMARY")
    print("═" * 50)
    
    total_success = 0
    total_failed = 0
    
    for layer, (success, failed) in all_results.items():
        total_success += len(success)
        total_failed += len(failed)
        status = "✅" if not failed else "⚠️"
        print(f"{status} {layer}: {len(success)} installed, {len(failed)} failed")
    
    print("─" * 50)
    print(f"Total installed: {total_success}")
    print(f"Total failed:    {total_failed}")


def print_next_steps():
    """Print what to build next."""
    print("\n\n🚀 YOUR STACK IS READY")
    print("═" * 50)
    print("""
WHILE-SLEEPING v2.0 — BUILD ORDER

Week 1 → Data Foundation
  └── 01_simulate_data.ipynb
       Generate 100M customer events
       Store in Delta Lake
       
Week 2 → Hypothesis Library  
  └── 02_hypothesis_library.ipynb
       Schema design
       Seed first hypotheses
       Human approval workflow

Week 3 → Agent Framework
  └── 03_agents.ipynb
       LangGraph orchestrator
       4 specialist agents
       MCP servers

Week 4 → Pain Detection
  └── 04_pain_detection.ipynb
       Real-time scoring
       Baseline comparison
       Anomaly detection

Week 5 → Self-Healing Dashboard
  └── 05_dashboard.py
       Streamlit UI
       Human guardrail layer
       WhatsApp alerts

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
MAPS TO BARCLAYS PRODUCTION:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
PySpark local    →  Databricks
Delta Lake local →  Delta Lake on S3  
Qwen local       →  Approved LLM
LangGraph        →  Copilot Studio
Streamlit        →  Teams + WhatsApp
MCP servers      →  Same (portable!) ✅
Hypothesis Lib   →  Same (portable!) ✅
GitLab personal  →  Barclays GitLab
""")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# MAIN EXECUTION
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
if __name__ == "__main__":
    
    print("""
╔══════════════════════════════════════════════════════════════════╗
║           WHILE-SLEEPING — STACK SETUP                          ║
║           Self-Healing Customer Journey System                   ║
║           Barclays Decision Intelligence                         ║
╚══════════════════════════════════════════════════════════════════╝
    """)
    
    all_results = {}
    
    # Install each layer
    s, f = install_layer("CORE DATA STACK",        core_packages,         "🗄️")
    all_results["Core Data Stack"] = (s, f)
    
    s, f = install_layer("STREAMING & EVENTS",     streaming_packages,    "⚡")
    all_results["Streaming & Events"] = (s, f)
    
    s, f = install_layer("AGENT FRAMEWORK",        agent_packages,        "🤖")
    all_results["Agent Framework"] = (s, f)
    
    s, f = install_layer("MCP NERVOUS SYSTEM",     mcp_packages,          "🧠")
    all_results["MCP Nervous System"] = (s, f)
    
    s, f = install_layer("AI & LLM",               ai_packages,           "✨")
    all_results["AI & LLM"] = (s, f)
    
    s, f = install_layer("ANALYTICS & DETECTION",  analytics_packages,    "🔍")
    all_results["Analytics & Detection"] = (s, f)
    
    s, f = install_layer("DASHBOARD",              dashboard_packages,    "📊")
    all_results["Dashboard"] = (s, f)
    
    s, f = install_layer("DATA QUALITY & LINEAGE", quality_packages,      "✅")
    all_results["Data Quality & Lineage"] = (s, f)
    
    s, f = install_layer("NOTIFICATIONS",          notification_packages, "🔔")
    all_results["Notifications"] = (s, f)
    
    s, f = install_layer("UTILITIES",              utility_packages,      "🛠️")
    all_results["Utilities"] = (s, f)
    
    # Create project structure
    create_project_structure()
    
    # Verify critical packages
    verify_stack()
    
    # Print summary
    print_summary(all_results)
    
    # Print next steps
    print_next_steps()
````
